```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder, StandardScaler,MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Load the dataset
# low_thresh = data["CO2_Emissions (in MT)"].quantile(0.33)
# high_thresh = data["CO2_Emissions (in MT)"].quantile(0.66)

# def categorize_pollution(value):
#     if value <= low_thresh:
#         return "Low"
#     elif value <= high_thresh:
#         return "Medium"
#     else:
#         return "High"\
data["Pollution_Severity"] = pd.cut(data["Energy_Recovered (in GWh)"], 
                                    bins=[data["Energy_Recovered (in GWh)"].min(), 
                                          data["Energy_Recovered (in GWh)"].median(), 
                                          data["Energy_Recovered (in GWh)"].quantile(0.75), 
                                          data["Energy_Recovered (in GWh)"].max()], 
                                    labels=["Low", "Medium", "High"],
                                    include_lowest=True)


# Apply category labels to CO2 emissions

# Encode target variable (Low -> 0, Medium -> 1, High -> 2)
le = LabelEncoder()
data["Pollution_Severity"] = le.fit_transform(data["Pollution_Severity"])

# Select relevant features for classification


data["pollution_total"] = data["Air_Pollution_Index"] + data["Water_Pollution_Index"]+data["Soil_Pollution_Index"]+data["Industrial_Waste (in tons)"]
scaler = MinMaxScaler()

data["pollution_total"] = scaler.fit_transform(data[["pollution_total"]])

features = ["Air_Pollution_Index", "Water_Pollution_Index", "Soil_Pollution_Index", "Industrial_Waste (in tons)","CO2_Emissions (in MT)",
           "Plastic_Waste_Produced (in tons)","Yearly_pollution_trend","pollution_total"]
# Scale features to improve KNN performance
scaler = MinMaxScaler()
data[features] = scaler.fit_transform(data[features])
# data["pollution_total"] = scaler.fit_transform(data[features])

# Split into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data[features], data["Pollution_Severity"], test_size=0.2, random_state=42)

# Hyperparameter tuning: Find the optimal K value
param_grid = {"n_neighbors": list(range(1, 21))}
grid_search = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5, scoring="accuracy")
grid_search.fit(X_train, y_train)

# Get the best K value
best_k = grid_search.best_params_["n_neighbors"]
print(f"Optimal K: {best_k}")

# Train KNN with the best K value
knn = KNeighborsClassifier(n_neighbors=best_k,weights ='distance')
knn.fit(X_train, y_train)

# Predictions
y_pred = knn.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=le.classes_)

print(f"Accuracy: {accuracy:.2f}")
print("Confusion Matrix:\n", conf_matrix)
print("Classification Report:\n", class_report)

# Plot K values vs Accuracy
k_values = list(range(1, 21))
accuracy_scores = [grid_search.cv_results_["mean_test_score"][i] for i in range(len(k_values))]

plt.figure(figsize=(8, 5))
plt.plot(k_values, accuracy_scores, marker="o", linestyle="dashed", color="b")
plt.xlabel("Number of Neighbors (K)")
plt.ylabel("Accuracy")
plt.title("K-Value Optimization for KNN")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import adjusted_rand_score, silhouette_score
import scipy.cluster.hierarchy as sch

# Load and preprocess data
# features = ["Air_Pollution_Index", "Water_Pollution_Index", "Soil_Pollution_Index", "Industrial_Waste (in tons)"]
data["Pollution_Severity"] = pd.cut(data["Energy_Recovered (in GWh)"], 
                                    bins=[data["Energy_Recovered (in GWh)"].min(), 
                                          data["Energy_Recovered (in GWh)"].median(), 
                                          data["Energy_Recovered (in GWh)"].quantile(0.75), 
                                          data["Energy_Recovered (in GWh)"].max()], 
                                    labels=["Low", "Medium", "High"],
                                    include_lowest=True)

le = LabelEncoder()
data["Pollution_Severity"] = le.fit_transform(data["Pollution_Severity"])
data
```

```python

data["total"] = data["Air_Pollution_Index"] + data["Water_Pollution_Index"]+data["Soil_Pollution_Index"]+data["Industrial_Waste (in tons)"]
total = data["total"]
# Apply category labels to CO2 emissions

# Encode target variable (Low -> 0, Medium -> 1, High -> 2)

data["total"] = le.fit_transform(data["total"])
features = ["total","Pollution_Severity"]
# Scale features
scaler = StandardScaler()
data[features] = scaler.fit_transform(data[features])

# Hierarchical Clustering - Dendrogram
plt.figure(figsize=(10, 5))
linkage_matrix=sch.linkage(data[features], method="ward")
sch.dendrogram(linkage_matrix, labels=data["Country"].values, leaf_rotation=45,truncate_mode="level",p=3)

import numpy as np

# Extract cluster distances
distances = linkage_matrix[:, 2]

# Find the largest gap in distances
gaps = np.diff(distances)
optimal_clusters = np.argmax(gaps) + 1  # +1 to adjust index

print(f"Suggested number of clusters: {optimal_clusters}")


plt.title("Dendrogram for Agglomerative Clustering")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distances")
plt.show()

# Perform Agglomerative Clustering
agg_clustering = AgglomerativeClustering(n_clusters=3, metric="euclidean", linkage="ward")
data["Cluster"] = agg_clustering.fit_predict(data[features])

# Evaluate Cluster Assignments
ari_score = adjusted_rand_score(data["Pollution_Severity"], data["Cluster"])
silhouette = silhouette_score(data[features], data["Cluster"])

print(f"Adjusted Rand Index (ARI): {ari_score:.2f}")  # Higher is better
print(f"Silhouette Score: {silhouette:.2f}")  # Closer to 1 is better

# Visualize the Clusters
plt.scatter(data["Air_Pollution_Index"], data["Water_Pollution_Index"], c=data["Cluster"], cmap="viridis")
plt.xlabel("Air Pollution Index")
plt.ylabel("Water Pollution Index")
plt.title("Agglomerative Clustering Results")
plt.colorbar(label="Cluster")
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# features = ["Air_Pollution_Index", "Water_Pollution_Index", "Soil_Pollution_Index", "Industrial_Waste (in tons)"]
# X = data[features]
data["total"] = data["Air_Pollution_Index"] + data["Water_Pollution_Index"]+data["Soil_Pollution_Index"]+data["Industrial_Waste (in tons)"]
data["total"] = le.fit_transform(data["total"])

x = data[["total"]]
y = data["Pollution_Severity"]


# 🔹 Train-Test Split (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Build Neural Network model

model = Sequential([
    Dense(16,activation='relu',input_shape=(X_train.shape[1],)),
    Dense(8, activation='relu'),
    Dense(1, activation = 'sigmoid')
])
    
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train model
model.fit(X_train, y_train, epochs=50, batch_size=2, verbose=1, validation_data=(X_test, y_test))

# Evaluate model
loss, accuracy = model.evaluate(X_test, y_test)
# print(f"Test Accuracy: {accuracy * 100:.2f}%")

# Predict probabilities for the test set
y_pred_prob= model.predict(X_test)
y_train_prob= model.predict(X_train)


# Apply threshold to classify as Fast (0) or Delayed (1)
y_pred = (y_pred_prob > 0.5).astype(int)

y_train = (y_train_prob > 0.5).astype(int)
```
